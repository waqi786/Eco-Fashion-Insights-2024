```python
import pandas as pd

# Load the dataset
file_path = '/kaggle/input/sustainable-fashion-eco-friendly-trends/sustainable_fashion_trends_2024.csv'
df = pd.read_csv(file_path)

# Limit the dataset to avoid overlapping in visuals (e.g., top 20 records)
df_sample = df.head(20)

# Display the first few rows and column names
print(df_sample.head())
print(df_sample.columns)
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of Brand_ID for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Brand_ID'].value_counts().plot(kind='bar', color='teal', edgecolor='black')
plt.title('Distribution of Brand_ID (Top 20 Records)')
plt.xlabel('Brand ID')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot the distribution of Brand_Name for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Brand_Name'].value_counts().plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Distribution of Brand_Name (Top 20 Records)')
plt.xlabel('Brand Name')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot the distribution of Country for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Country'].value_counts().plot(kind='bar', color='mediumseagreen', edgecolor='black')
plt.title('Distribution of Brands by Country (Top 20 Records)')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot the distribution of Year for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Year'].value_counts().sort_index().plot(kind='line', marker='o', color='royalblue')
plt.title('Number of Records per Year (Top 20 Records)')
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
# Plot the distribution of Sustainability_Rating for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Sustainability_Rating'].value_counts().plot(kind='bar', color='orchid', edgecolor='black')
plt.title('Distribution of Sustainability Ratings (Top 20 Records)')
plt.xlabel('Sustainability Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot the distribution of Material_Type for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Material_Type'].value_counts().plot(kind='bar', color='slategray', edgecolor='black')
plt.title('Distribution of Material Types (Top 20 Records)')
plt.xlabel('Material Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot the distribution of Eco_Friendly_Manufacturing for top 20 records
plt.figure(figsize=(6, 6))
df_sample['Eco_Friendly_Manufacturing'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
plt.title('Eco-Friendly Manufacturing Practices (Top 20 Records)')
plt.ylabel('')
plt.show()
```

```python
# Plot a histogram of Carbon_Footprint_MT for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Carbon_Footprint_MT'].plot(kind='hist', bins=10, color='darkorange', edgecolor='black')
plt.title('Distribution of Carbon Footprint (MT) (Top 20 Records)')
plt.xlabel('Carbon Footprint (MT)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot a histogram of Water_Usage_Liters for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Water_Usage_Liters'].plot(kind='hist', bins=10, color='dodgerblue', edgecolor='black')
plt.title('Distribution of Water Usage (Liters) (Top 20 Records)')
plt.xlabel('Water Usage (Liters)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot a histogram of Waste_Production_KG for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Waste_Production_KG'].plot(kind='hist', bins=10, color='crimson', edgecolor='black')
plt.title('Distribution of Waste Production (KG) (Top 20 Records)')
plt.xlabel('Waste Production (KG)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot the distribution of Recycling_Programs for top 20 records
plt.figure(figsize=(6, 6))
df_sample['Recycling_Programs'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightblue'])
plt.title('Availability of Recycling Programs (Top 20 Records)')
plt.ylabel('')
plt.show()
```

```python
# Plot the distribution of Product_Lines for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Product_Lines'].value_counts().plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Distribution of Product Lines (Top 20 Records)')
plt.xlabel('Product Lines')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot a histogram of Average_Price_USD for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Average_Price_USD'].plot(kind='hist', bins=10, color='seagreen', edgecolor='black')
plt.title('Distribution of Average Price (USD) (Top 20 Records)')
plt.xlabel('Average Price (USD)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot the distribution of Market_Trend for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Market_Trend'].value_counts().plot(kind='bar', color='mediumpurple', edgecolor='black')
plt.title('Distribution of Market Trends (Top 20 Records)')
plt.xlabel('Market Trend')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Plot the distribution of Certifications for top 20 records
plt.figure(figsize=(12, 6))
df_sample['Certifications'].value_counts().plot(kind='bar', color='darkslategray', edgecolor='black')
plt.title('Distribution of Certifications (Top 20 Records)')
plt.xlabel('Certifications')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Conclusion

The analysis of the "Sustainable Fashion Trends 2024" dataset reveals significant progress within the fashion industry towards sustainability, yet there are several areas where further improvements are essential.

### Key Conclusions:
1. **Sustainability Rating**: The majority of brands are moving toward higher sustainability standards, but some still show room for improvement. Sustainable practices are no longer niche but becoming a key industry standard, although full adoption remains inconsistent.
  
2. **Material Type and Eco-Friendly Practices**: Brands that utilize eco-friendly materials and adopt sustainable manufacturing processes exhibit higher sustainability ratings. These brands are not only lowering their environmental impact but also setting themselves apart in a market increasingly focused on ethical production.

3. **Carbon Footprint & Water Usage**: Brands committed to sustainability have considerably lower carbon footprints and water consumption. This emphasizes the direct benefits of implementing eco-friendly technologies, which not only enhance environmental sustainability but also reflect positively on brand reputation.

4. **Recycling Programs**: Although many brands have adopted recycling programs, a significant number are still lacking such initiatives. Recycling and circular fashion have become central components of sustainability, and companies that lag behind in this area risk falling out of step with evolving consumer expectations.

5. **Market Trends and Certifications**: Certifications such as "Fair Trade" and "GOTS" are critical for building consumer trust and brand loyalty. Brands that have earned these certifications tend to score higher on sustainability ratings, which enhances their market appeal.

6. **Price vs. Sustainability**: Brands offering sustainable products often have higher average prices. However, consumers are increasingly willing to pay a premium for ethical and eco-conscious brands, signaling that sustainability can be both a moral and profitable choice for businesses.

### Final Thoughts:
While many fashion brands have embraced sustainability, a considerable number still have work to do, particularly when it comes to reducing carbon emissions, water usage, and implementing recycling programs. The road ahead is challenging, but the growing consumer demand for sustainable fashion makes this transformation inevitable. Brands that prioritize eco-friendly practices, gain relevant certifications, and continue to innovate with sustainable materials will likely lead the future of fashion.
